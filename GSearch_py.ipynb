```
import pandas as pd
from googlesearch import search

# Function to perform a Google search and return the first result URL
def google_search(query):
    try:
        search_results = list(search(query, num_results=1))
        if search_results:
            return search_results[0]
        else:
            return None
    except Exception as e:
        return str(e)

# Load the Excel file
file_path = 'ela.xlsx'
df = pd.read_excel(file_path)

# Define the name of the new column to store the search results
new_column_name = 'Search Results'

# Iterate through each row and perform a Google search for each word
search_results = []
for index, row in df.iterrows():
    search_query = ' '.join(str(row['word']))  # Replace 'Your_Column_Name' with the actual column name
    result = google_search(search_query)
    search_results.append(result)

# Add the search results as a new column in the DataFrame
df[new_column_name] = search_results

# Save the DataFrame back to the Excel file
df.to_excel(file_path, index=False)

print(f'Search results added to the "{new_column_name}" column.')
```
